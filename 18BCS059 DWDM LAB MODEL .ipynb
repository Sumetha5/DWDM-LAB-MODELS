Apriori Algorithm

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from apyori import apriori
```

```python
data = pd.read_csv('retail_dataset.csv')
data
```

```python
data.shape
```

```python
data.describe()
```

```python
records = []
for i in range(0,315):
    records.append([str(data.values[i,j]) for j in range(0,7)])
```

```python
for sets in records:
    print(sets)
```

```python
association_rules = apriori(records, min_support = 0.005, min_confidence = 0.7, min_lift = 1, min_length = 2)
```

```python
association_results = list(association_rules)
```

```python
print(len(association_results))
```

```python
print(association_results[0])
```

```python
for item in association_results:
    pair = item[0]
    items = [x for x in pair]
    print("Rule :"+ str(item[0]) + "->" + str(item[1]))
    print("Support : {}".format(item[1]))
    print("Confidence : {}".format(item[2][0][2]))
    print("Lift : {}".format(item[2][0][3]))
    print("\n------------------------------------------------\n")
```
